'Country name'lerinde 'Regional indicator' olmayanlar için tek tek eşleşme yapıldı.

```python
data.head()
#Yeni region'lar
```

```python
all_countries = data["Country name"].value_counts().reset_index()["Country name"].tolist()
all_countries_pop = datanufus["Country Name"].value_counts().reset_index()["Country Name"].tolist()

del_cou = []
for x in all_countries:
    if x not in all_countries_pop:
        del_cou.append(x)
del_cou
```

Üsttekiler, ana listede olup nüfus listesinde olayan ülkelerdir.

```python
data.columns
```

```python
population_df = data[['Log GDP per capita', 'Life Ladder', 'Country name', 'year', 'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',"region",'Perceptions of corruption']].copy()
population_df.head()
```

```python
#silinen değerler
population_df = population_df[~population_df["Country name"].isin(del_cou)]
population_df = population_df[~population_df.year.isin([2006,2005,2007,2018,2019,2020,2021])]
population_dict = {x:{} for x in range(2008,2018)}
for i in range(len(datanufus)):
    if(datanufus["Year"][i] in range(2008,2018)):
        population_dict[datanufus["Year"][i]][datanufus["Country Name"][i]] = datanufus["Count"][i]
```

```python
population_dict
#her bir ülkenin nüfusu
```

```python
population = []
for i in population_df.index:
    population.append(population_dict[population_df["year"][i]][population_df["Country name"][i]])
population_df["population"] = population

population_df.head()
```

```python
fig = px.scatter(population_df, 
                 x = "Log GDP per capita",
                 y = "Life Ladder",
                 animation_frame = "year",
                 animation_group = "Country name",
                 size = "population",
                 template = "plotly_white",
                 color = "region", 
                 hover_name = "Country name", 
                 size_max = 60)
fig.update_layout(title = "Ülkelerin gelirlerine göre mutluluk oranı")
fig.show()
```

Genele göre incelersek yıllar içinde Günlük GSYİH'sı artan ülkelerde mutluluk oranının da arttığını söyleyebiliriz.

```python
fig = px.scatter(population_df, 
                 x = "Freedom to make life choices",
                 y = "Life Ladder",
                 animation_frame = "year",
                 animation_group = "Country name",
                 size = "population",
                 template = "plotly_dark",
                 color = "region", 
                 hover_name = "Country name", 
                 size_max = 60)
fig.update_layout(title = "Ülkelerin özgürlüklerine göre mutluluk oranı")
fig.show()
```
