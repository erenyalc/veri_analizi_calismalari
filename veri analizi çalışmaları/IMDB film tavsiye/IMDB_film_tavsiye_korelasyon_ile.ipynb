```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("users.data", sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
data
```

```python
data2 = pd.read_csv("movie_id_titles.csv")
data2
```

İkisinde de 'item_id' var. item_id'ye göre birleştireceğiz.

```python
data = pd.merge(data, data2, on='item_id')
data
```

```python
pivot_tablo = data.pivot_table(index='user_id', columns='title', values='rating')
pivot_tablo
```

Bir kullanıcının hangi filme kaç puan verdiğini gösteren bir pivot tablodur.

```python
starwars = pivot_tablo['Star Wars (1977)']
starwars
type(starwars)
```

Star Wars'u izleyen kişilere izleyebilecekleri diğer filmleri sunalım.

```python
korelasyon = pivot_tablo.corrwith(starwars)
korelasyon
```

Star wars 1977 filmi ile diğer filmler arasındaki puan korelasyonu incelendi

```python
kor = pd.DataFrame(korelasyon, columns=['Correlation'])
kor.dropna(inplace = True)
kor.head()
```

```python
kor.sort_values('Correlation', ascending=False).head()
```

```python
data.drop((['timestamp']), axis=1)
```

```python
ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
ratings.sort_values('rating', ascending=False)
```

```python
ratings['oysayisi'] = pd.DataFrame(data.groupby('title')['rating'].count())
ratings.head()
```

```python
ratings.sort_values('oysayisi', ascending=False).head()
```

```python
kor = kor.join(ratings['oysayisi'])
kor.head()
```

oy sayısı ile korelasyon tabloları birleştirildi. 100'den az oy alan filmler dahil edilmeyecek. Tahmini bozmaması adına.

```python
kor[kor['oysayisi']>100].sort_values('Correlation', ascending=False).head()
```
